```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris') #This line of code load a dataset called iris which are present in seaborn
#NOTE: You can load a dataset from external source as well
```

```python
iris
```

```python
#Construct a graph to see the relation between sepal_length and sepal_width
sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
#Construct a graph to see the relation between sepal_length and petal_length
sns.scatterplot(x = iris.sepal_length, y = iris.petal_length)
```

```python
#Construct a graph to see the relation between petal_length and petal_width
sns.scatterplot(x = iris.petal_width , y = iris.petal_length )
```

### displot()

```python
sns.displot(iris['sepal_length']) #distribution plot is equivalent to Histogram
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

### scatterplot()

```python
#Total bill vs Tip
sns.scatterplot(x = tips.total_bill, y = tips.tip)
```

```python
tips.head()
```

```python
#Finding total number of smokers and non-smokers
tips['smoker'].value_counts()
```

```python
#Relational plot
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, hue = 'smoker')
#Here hue = 'smoker' will segregate with different colours on the basis of smoker or not smoker
```

```python
tips.head()
```

```python
tips['size'].value_counts()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, hue = 'size' )
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, style = 'size' )
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, hue = 'size', style = 'size' )
```

```python
tips.head()
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, hue ='size', style = 'time' )
```

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, hue ='size', style = 'day' )
```

### catplot()

```python
#Categorical plot : day vs total_bill
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
tips.head()
```

```python
#Categorical plot : smoker vs total_bill
sns.catplot(x = 'smoker', y = 'total_bill', data = tips)
```

```python
#Categorical plot : smoker vs tip
sns.catplot(x = 'smoker', y = 'tip', data = tips)
```

### jointplot()

```python
sns.jointplot(x = tips.total_bill, y = tips.tip) #This will show scatterplot as well as histogram
```

### pairplot()

```python
sns.pairplot(tips)
```
