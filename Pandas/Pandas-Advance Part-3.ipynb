#### Let us see how we can remove a column 
>> Row  are known as axis = 0 in pandas
>>> Column are known as axis = 1 in pandas

So, if we want to to drop any column then we have to pass column name along with axis = 1 

>> Code:
df.drop('PassengerId', axis = 1)

But problem with this line of code is this will not delete 'PassengerId' column permanently as we can see below when be print df

>>Solution: df.drop('PassengerId', axis = 1, inplace = true)
>>>> 2nd solution can be we can overwrite ex: 
df = df.drop('PassengerId', axis = 1)
         

```python
df.drop('PassengerId', axis = 1, inplace = True)
```

#### What is the difference between merge () and concat () in Pandas?
3 Key Differences Between Merge and Concat Functions of ...
>>Concat function concatenates dataframes along rows or columns. We can think of it as stacking up multiple dataframes. 

>>Merge combines dataframes based on values in shared columns. Merge function offers more flexibility compared to concat function because it allows combinations based on a condition

### Left 

```python
pd.merge(data1,data2,how = 'left')
```

```python
data1
```

### JOIN

```python
data1 = pd.DataFrame({'Key1':[1,2,4,5,6],
                      'Key2':[4,5,6,7,8],
                      'Key3':[3,4,5,6,7]},
                      index = ['a','b','c','d','e']
                      
    
)
```

```python
data2 = pd.DataFrame({'Key11':[1,2,4,5,6],
                      'Key22':[4,5,6,7,8],
                      'Key33':[3,4,5,6,7]},
                     index = ['a','b','h','i','j']
                      
    
)
```

```python
data1
```

```python
data2
```

#### Left Join

```python
data1.join(data2)
```

#### Right Join

```python
data1.join(data2,how = 'right')
```

#### Inner Join

```python
data1.join(data2,how = 'inner')
```

#### Outer Join

```python
data1.join(data2,how = 'outer')
```

#### Cross Join

```python
data1.join(data2,how = 'cross')
```

### Lets see a interesting operations on Fare using apply() function.

```python
#Here Fare is given in euro Our task is to convert it into Rupee 
#and create a new column with the name Fare_Rupee and store that 
#new Fare after conversion in that column
#Solution:
```

```python
df['Fare_INR'] = df['Fare'].apply(lambda x : x*80) #Here lambda us our insider function
```

```python
df.head()
```

```python
#Lets see how we can do the same with outersider function 
def euro_inr(x):
    return x*80

df['Fare_INR'] = df['Fare'].apply(euro_inr)
```

```python
df
```

```python
df.head()
```

```python
#Here our tast is to store the length of Name which are present in Name column in a New column
```

```python
df['Name_len'] = df['Name'].apply(len)
```

```python
df.head()
```

```python
#Now our task is to create a column which assign a category
#IF FARE >= 10 : mid
#IF FARE < 10 : Cheap
#else : High
```

```python
#Solution:
```

```python
def cat_fare(x):
    if x<10:
        return "cheap"
    elif x >= 10 and x < 20:
        return "mid"
    else:
        return 'high'
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
